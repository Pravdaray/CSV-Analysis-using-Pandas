```python
#change csv to pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('a1.csv', delimiter=',')
df.index.names = ['Number of Papers']

sns.barplot(x=df.index,y=df.Task)
plt.show()
plt.savefig('paper_dist.png', dpi = 300)
```

```python
#df['Year', 'Task'].value_counts().plot(kind='barh')
df.plot(x = 'Year', y = ['Task', 'ROI'])
plt.show
```

```python
#merging multiple labels
import matplotlib.pyplot as plt
df_segmentation = df[df["Task"] == "Segmentation"]
df_detection = df[df["Task"] == "Detection"]
df_classification = df[df["Task"] == "Classification"]
sns.set(style="white", color_codes=True)
sns.distplot(df_segmentation['Year'], kde= False,color = "Green", label = "Segmentation", bins = 12)
sns.distplot(df_detection['Year'], kde= False, color = "Red" , label = "Detection", bins = 12)
sns.distplot(df_classification['Year'], kde= False, color = "Blue" , label = "Classification", bins = 12)
#plt.legend(prop={'size': 20})
plt.title('Number of Papers')
plt.xlabel('Year')
plt.ylabel('Number of papers')
plt.savefig('number_of_paper.png', dpi=300)
```

```python
#number of papers only based on task
plt.figure(figsize=(8,4))
sns.countplot(data=df, x = 'Task');
```

```python
import pandas as pd
df2 = pd.read_csv("review_table.csv")
df2.head()
```

```python
#task and roi
ax = sns.countplot(x="Task", hue="ROI", data=df2, palette="rainbow")
plt.savefig('task_roi.png', dpi=300)
```

```python
#modality and year analysis
import matplotlib.pyplot as plt

df_ct = df2[df2["Modalality"] == "CT"]
df_us = df2[df2["Modalality"] == "US"]
df_mr= df2[df2["Modalality"] == "MR"]
sns.set(style="white", color_codes=True)
sns.distplot(df_ct['Year'], kde= False,color = "Green", label = "CT", bins = 15)
sns.distplot(df_us['Year'], kde= False, color = "Red" , label = "US", bins = 15)
sns.distplot(df_mr['Year'], kde= False, color = "Blue" , label = "MR", bins = 15)
#plt.legend(prop={'size': 20})
plt.title('Number of Papers')
plt.xlabel('Year')
plt.ylabel('Number of papers')
plt.savefig('number_of_paper.png', dpi=300)
```
